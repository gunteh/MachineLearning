```python
import pandas as pd
```

```python
file_path = "~/Desktop/ethics_AI/Advertisers_condescended - 2nd iteration.csv"
```

```python
df = pd.read_csv(file_path)
```

```python
print(df.head())
```

```python
df = df.drop_duplicates(subset=['Advertiser', category_column])
```

```python
category_column = df.columns[1] # Category
print(df[category_column].value_counts())
print(df[category_column].value_counts().to_string())
```

```python
print(df['Advertiser'].value_counts())
```

```python
# Filter advertiser results based on categories with more than 30 advertisers,
# then perform stratified random sampling per category.

advertiser_category_counts = df[category_column].value_counts()
validation_category_list = advertiser_category_counts[advertiser_category_counts > 30].index.tolist() 
# Ensuring statistical significance by selecting categories with sufficient samples.

# Filtered DataFrame containing only categories with more than 30 advertisers.
filtered_advertiser_dataframe = df[df[category_column].isin(validation_category_list)]

# Stratified sampling to obtain randomized advertisers from each qualifying category.
# Create an empty list to store sampled advertiser data.
stratified_sampled_advertisers = []

# Grouping the  filterd DataFrame by category type.
grouped_advertiser_dataframe = filtered_advertiser_dataframe.groupby(category_column)

# Loop through each advertiser category and randomly select 20 rows per category.
for advertiser_category_name, category_dataframe in grouped_advertiser_dataframe:
    # Randomly select 20 advertisers per category.
    # Set a fixed seed value for  reproducibility.
    sampled_advertiser_subset = category_dataframe.sample(n=20, random_state=42)

    # Append the sampled subset to the master list.
    stratified_sampled_advertisers.append(sampled_advertiser_subset)

# Merge all sampled groups into a single DataFrame.
final_stratified_advertiser_dataframe = pd.concat(stratified_sampled_advertisers, ignore_index=True)
```

```python
print(final_stratified_advertiser_dataframe.head())
print(final_stratified_advertiser_dataframe[category_column].value_counts())
```

```python
advertiser_column = final_stratified_advertiser_dataframe.columns[0] #Advertiser
print(final_stratified_advertiser_dataframe[advertiser_column].value_counts())
```

```python
final_stratified_advertiser_dataframe.to_csv("Advertisers_stratified_sampling.csv", index=False)
```
